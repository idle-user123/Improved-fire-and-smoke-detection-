```python
# Create the dataset.yaml file
yaml_content = """
train: /content/dataset/images/train
val: /content/dataset/images/val
test: /content/dataset/images/test

nc: 2
names: ['fire', 'smoke']
"""

with open('/content/dataset.yaml', 'w') as file:
    file.write(yaml_content)

print("dataset.yaml created successfully!")
```

```python
from ultralytics import YOLO

# Load YOLOv8n model
model = YOLO('yolov8n.pt')  # Pre-trained YOLOv8n weights

# Train the model using your dataset.yaml
model.train(data='/content/dataset.yaml', epochs=5, imgsz=640, batch=16)
```
